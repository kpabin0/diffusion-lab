```python
import torch
import torch.distributions as D

from diflab.models import MLP
from diflab.utils import customProb2D
import torch.nn.functional as F
```

```python
e = torch.randn(1000, 2)
loc = torch.tensor([0.0, 0.0])
scale = torch.tensor([1.0, 1.0])
probs = customProb2D(loc, scale)
z = probs.sample(1000)
model = MLP()
```

```python
def loss_fn(y, z, e):
    loss = F.mse_loss(y, (z - e))
    return loss
```

```python
optim = torch.optim.Adam(MLP().parameters())
epochs = 20000
for epoch in range(epochs):
    t = torch.sigmoid(torch.randn(1000, 1))
    x = t*z + (1-t)*e
    loss = loss_fn(model(x), z, e)
    optim.zero_grad()
    loss.backward()
    optim.step()
    if (epoch + 1) % 500 == 0:
        print(f"Epoch :[{epoch +1}/{epochs}] completed. Loss = {loss}")
```
