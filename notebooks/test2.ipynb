```python
import torch
import seaborn as sns

from diflab.utils import customProb2D
from diflab.models.mlp import MLP
```

```python
import matplotlib.pyplot as plt
```

```python
x_0 = torch.randn(1000, 2)
loc = torch.tensor([0.0, 0.0])
scale = torch.tensor([1.0, 1.0])
probs = customProb2D(loc, scale)
x = probs.sample(1000)
```

```python
N = 5
x_i = torch.zeros_like(x)

fig, axes = plt.subplots(nrows=1, ncols=N+1, figsize=(20, 3))
for i in range(0, N+1):
    x_i = (1 - i/N) * x_0 + i/N * x
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    sns.kdeplot(x=x_i[:,0], y=x_i[:,1], fill=True, levels=10, ax=axes[i])
```
